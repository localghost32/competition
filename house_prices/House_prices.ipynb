```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
train.head(5)
```

```python
train.info()
```

### Data prepare

```python
y_train = train.SalePrice
train.drop('SalePrice', axis=1, inplace = True)
```

```python
train.columns == test.columns
```

```python
train['is_test'] = 0
test['is_test'] = 1
df = pd.concat([train, test])
```

```python
df.info()
```

```python
df2 = df.dtypes
# columns with value's type "object"
list_of_obj_col = df2[df2=='object'].index.to_list()
print(len(list_of_obj_col))
```

```python
df_dummies = pd.get_dummies(df, columns=list_of_obj_col)
```

```python
df_dummies.isnull().sum()
```

```python
X_train = df_dummies[df_dummies.is_test==0].drop('is_test', axis=1)
X_test = df_dummies[df_dummies.is_test==1].drop('is_test', axis=1)
```

```python
X_train.head()
```

### Fill missing values

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
```

```python
imputer
```

```python
imputer.fit(X_train)
```

```python
X_train_imputed = imputer.transform(X_train)
X_train_imputed = pd.DataFrame(X_train_imputed, columns=X_train.columns)
```

```python
X_train_imputed.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train_imputed)
```

```python
X_train_imputed_scaled = scaler.transform(X_train_imputed)
X_train_imputed_scaled = pd.DataFrame(X_train_imputed_scaled,columns=X_train.columns)
```

```python
X_train_imputed_scaled.head(10)
```

#### Regression model

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression()
```

```python
clf.fit(X_train_imputed_scaled, y_train)
```

```python
imputer.fit(X_test)
scaler.fit(X_test)
y_pred = clf.predict(scaler.transform(imputer.transform(X_test)))
```

```python
y_pred
```

```python
result = X_test[['Id']]
result
```

```python
result['SalePrice'] = pd.DataFrame(y_pred)
```
